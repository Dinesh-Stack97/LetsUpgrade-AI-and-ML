## Hypotheis 1
## Null Hypothesis: There is no significant  correlation between Attrition and Monthly Income.
## Alternanative Hypothesis: There is significant correlation between  Attrition and Monthly Income. 

```python
import pandas as pd

```

```python
from sklearn import preprocessing

Label_encoder = preprocessing.LabelBinarizer() 
dataset1["Attrition"] = Label_encoder.fit_transform (dataset1["Attrition"])

from scipy.stats import pearsonr

stats,p = pearsonr(dataset1.Attrition, dataset1.MonthlyIncome)
print(stats,p)
```

## Inference

__since p = 0.038427<0.05 accept Alternative hypothesis__
__i.e "There is significant correlation between  Attrition and Monthly Income".__

# Hypotheis 2
## Null Hypothesis: There is no significant  correlation between Attrition and EducationField.
## Alternanative Hypothesis: There is significant correlation between  Attrition and EducationField.

```python
from sklearn import preprocessing

Label_encoder = preprocessing.LabelBinarizer()
dataset1["Attrition"] = Label_encoder.fit_transform(dataset1["Attrition"])
dataset1["EducationField"] = Label_encoder.fit_transform(dataset1["EducationField"])

from scipy.stats import pearsonr

stats,p = pearsonr(dataset1.Attrition, dataset1.EducationField)
print(stats,p)
```

__since p = 1.113309814e-09<0.05 i.e "There is significant correlation between  Attrition and EducationField".__

# Hypotheis 3
## Null Hypothesis: There is no significant  correlation between Attrition and Age.

```python
from sklearn import preprocessing

Label_encoder = preprocessing.LabelBinarizer()
dataset1["Attrition"] = Label_encoder.fit_transform(dataset1["Attrition"])

from scipy.stats import pearsonr

stats, p = pearsonr(dataset1.Attrition,dataset1.Age)
print("statistical value = ",stats )
print("probability value = ",p)
```
