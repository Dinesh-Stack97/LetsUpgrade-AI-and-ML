# Wilcoxon Sign Test

# Friedman Test
## used to compare more than two paired samples

__null hypothesis: There is differnce in attretion comparing with gender, job role ,salary__

```python
from sklearn import preprocessing

Label_encoder = preprocessing.LabelBinarizer()
dataset1["Attrition"] = Label_encoder.fit_transform(dataset1["Attrition"])
dataset1["Gender"] = Label_encoder.fit_transform(dataset1["Gender"])
dataset1["JobRole"] = Label_encoder.fit_transform(dataset1["JobRole"])


from scipy.stats import friedmanchisquare

z1=dataset1.Attrition

z2=dataset1.Gender

z3=dataset1.JobRole


stat, p=friedmanchisquare(z1, z2, z3)
print(stat, p)
```
